```python
from langchain.chains import LLMChain
from langchain.prompts.chat import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate
)
from langchain.memory import ConversationBufferMemory

# Create a new agent class that uses the LLMChain
class Agent:
    def __init__(self, name, system_prompt, llm):
        self.name = name
        self.llm = llm
        self.system_prompt = system_prompt
        # Create a chat prompt that includes 'conversation_history'
        self.chat_prompt = ChatPromptTemplate.from_messages([
            SystemMessagePromptTemplate.from_template(self.system_prompt),
            # Include the conversation history and the last message in the prompt
            HumanMessagePromptTemplate.from_template("{conversation_history}\n{input}")
        ])
        self.chain = LLMChain(
            llm=self.llm,
            prompt=self.chat_prompt,
            verbose=False
        )

    def respond(self, input_text, conversation_history):
        response = self.chain.predict(
            input=input_text,
            conversation_history=conversation_history
        )
        return response
```

```python
# LLM configuration for the agent
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7,
    openai_api_key="")
```

```python
# Define all agents in the system

# Customer Proxy Agent
customer_proxy_agent = Agent(
    name="Customer_Proxy",
    system_prompt="You are a Customer Proxy Agent. Your role is to communicate customer's requirements to the Project Manager.",
    llm=llm
)

# Project Manager Agent
project_manager_agent = Agent(
    name="Project_Manager",
    system_prompt=(
        "You are a Project Manager working with an Agile Scrum team buiulding a MVP. Your responsibilities include gathering project requirements from stakeholders, "
        "coordinating with team members across all disciplines, defining the project scope, and monitoring progress across sprints. "
        "You are responsible for developing the overall project plan, defining timelines and delivery milestones, managing risks and dependencies, and ensuring that the project "
        "is delivered on time and within scope. "
        "You work closely with roles like Business Analysts, Developers, QA Engineers, Designers, and others to gather input, resolve blockers, "
        "synchronize efforts, and report overall project health to stakeholders. "
        "You consolidate estimates and help determine the number of sprints needed for the MVP based on team velocity and complexity estimates from other agents."
    ),
    llm=llm
)

# Business Analyst Agent
business_analyst_agent = Agent(
    name="Business_Analyst",
    system_prompt=(
        "You are a Business Analyst in a Scrum team. Your role is to analyze customer needs, break them down into high-level Epics and detailed User Stories, "
        "and ensure alignment with business goals. You collaborate with the team to refine the backlog and provide sizing input that helps inform planning. "
        "Your input supports sprint planning, release forecasting, and roadmap development."
    ),
    llm=llm
)

# System Architect Agent
architect_agent = Agent(
    name="System_Architect",
    system_prompt=(
        "You are a System Architect in a Scrum team. Your role is to define and evolve the system architecture in alignment with user stories, UI/UX designs, and business goals. "
        "You identify technical components, assess architectural complexity, and advise on the effort and sequencing of implementation. "
        "Your input helps determine which sprints architectural tasks fall into and flags any system-level dependencies."
    ),
    llm=llm
)

# Developer Agent
developer_agent = Agent(
    name="Developer",
    system_prompt=(
        "You are a Developer in a Scrum team. Your role is to implement features based on user stories and designs. "
        "You assess the complexity of development tasks, collaborate with the team on estimating story points, and share input on technical feasibility and risk. "
        "You help define velocity by contributing to the team's understanding of how much work can be completed per sprint."
    ),
    llm=llm
)

# QA Engineer Agent
qa_engineer_agent = Agent(
    name="QA_Engineer",
    system_prompt=(
        "You are a QA Engineer in a Scrum team. You define the testing strategy and estimate the effort required for functional, integration, and regression testing. "
        "You review user stories for testability, plan automation coverage, and help identify which stories are higher risk from a quality perspective. "
        "Your feedback helps determine the test load per sprint and where test-driven development may be required."
    ),
    llm=llm
)

# UI/UX Designer Agent
designer_agent = Agent(
    name="UI_UX_Designer",
    system_prompt=(
        "You are a UI/UX Designer in a Scrum team. You provide wireframes, user flows, and design recommendations for product features based on user stories. "
        "You provide input on design complexity and collaborate with developers to ensure designs are feasible within sprint timelines. "
        "Your input helps inform sprint planning and identify design dependencies across Epics."
    ),
    llm=llm
)

# DevOps Engineer Agent
devops_engineer_agent = Agent(
    name="DevOps_Engineer",
    system_prompt=(
        "You are a DevOps Engineer in a Scrum team. You are responsible for setting up and maintaining CI/CD pipelines, deployment environments, and monitoring tools. "
        "You assess the effort needed to integrate these systems, flag environment setup dependencies, and suggest which sprints may require DevOps milestones."
    ),
    llm=llm
)

# Security Engineer Agent
security_engineer_agent = Agent(
    name="Security_Engineer",
    system_prompt=(
        "You are a Security Engineer in a Scrum team. Your role is to ensure secure architecture, coding, and deployment practices are followed across sprints. "
        "You review Epics and technical designs for potential risks, propose mitigations, and help determine where to embed security tasks in the sprint plan."
    ),
    llm=llm
)

# Technical Writer Agent
technical_writer_agent = Agent(
    name="Technical_Writer",
    system_prompt=(
        "You are a Technical Writer in a Scrum team. You create user-facing documentation, API references, and onboarding materials based on feature development. "
        "You provide estimates on documentation workload per sprint and identify documentation tasks that should align with sprint deliveries and release milestones."
    ),
    llm=llm
)

# Scrum Master Agent
scrum_master_agent = Agent(
    name="Scrum_Master",
    system_prompt=(
        "You are a Scrum Master in a Scrum team. You facilitate sprint planning, reviews, and retrospectives, and support team alignment and agile best practices. "
        "You work closely with all team members to determine sprint duration (e.g., 2 weeks), estimate team velocity, and calculate how many sprints are likely needed to deliver the MVP."
    ),
    llm=llm
)


```

```python
# Define the conversation prompts from one agent to another

project_manager_prompts = {
    
    "Business_Analyst": (
        "We’ve received the customer requirements for the food delivery application. Please break them down into high-level Epics and detailed User Stories "
        "with clear acceptance criteria. Ensure the stories are logically grouped and aligned with business goals. "
        "Suggest a priority order, and provide early insights on the sizing and complexity of these stories to support planning. "
        "Also include your view on how many Epics and Stories we might be working with."
    ),

    "Scrum_Master": (
        "The Business Analyst has shared the structured backlog. Please help us define the overall sprint structure for this project. "
        "Recommend a sprint duration that suits the team, suggest a total number of sprints required based on typical velocity and story points suggested, "
        "and help define how we’ll organize sprint ceremonies including planning, stand-ups, reviews, and retrospectives and include that in the efforts"
    ),

    "UI_UX_Designer": (
        "Please review the initial user stories and identify design needs and complexity across each Epic. "
        "Share how many screens or UI components you anticipate designing accoss all the epics/user stories "
        "Highlight any dependencies that may impact timelines, and suggest how we can collaborate efficiently during sprint execution."
        "Finally suggest your efforts in storypoints required for each user story"
    ),

    "System_Architect": (
        "Based on the prioritized backlog and initial designs, please outline your proposed architecture for the system. "
        "Identify core components, data flow, and potential risks or bottlenecks. "
        "Provide a high-level timeline of when key architectural tasks should be completed across the upcoming sprints."
        "Finally suggest your efforts in storypoints required for each user story"
    ),

    "Developer": (
        "With the architecture and designs ready, please evaluate the user stories planned for early sprints. "
        "Comment on technical feasibility, complexity, and how the work might be distributed across sprints. "
        "Estimate how long the initial implementation phase might take and provide your view on the sprint-by-sprint development plan."
    ),

    "Security_Engineer": (
        "Please review the architecture and planned features from a security standpoint. "
        "Identify potential risks, propose security checkpoints during development, and estimate when security tasks should occur across the sprint cycle. "
        "Also suggest how we can embed secure coding and testing practices from the start."
        "Finally suggest your efforts in storypoints required for each user story"
    ),

    "DevOps_Engineer": (
        "As development gets underway, please define what DevOps infrastructure we’ll need (CI/CD, environments, pipelines, etc.). "
        "Estimate how long setup will take, and recommend how DevOps tasks can be distributed or automated throughout the sprints. "
        "Highlight any early blockers or critical dependencies to consider."
        "Finally suggest your efforts in storypoints required for each user story"
    ),

    "QA_Engineer": (
        "Please review the backlog and planned sprint features to define your initial QA strategy. "
        "Let us know your preferred approach for test case development, automation, and test cycles across sprints. "
        "Estimate when core QA work should begin and how it will scale as feature complexity increases."
        "Finally suggest your efforts in storypoints required for each user story"
    ),

    "Technical_Writer": (
        "As we start planning sprint work, please review the backlog and share your thoughts on documentation needs for end-users and internal teams. "
        "Provide a breakdown of what docs will be needed per feature or Epic, and suggest how documentation tasks can be aligned with sprint delivery milestones."
        "Finally suggest your efforts in storypoints required for each user story"
    ),
    "Project_Manager": (
        "You have received inputs from all Agile team members regarding Epics, User Stories, estimates, sprint duration, and timeline. "
        "Based on this, develop a complete project plan. Include the total number of sprints, a sprint-by-sprint delivery roadmap, and overall project duration. "
        "Ensure risks, dependencies, and cross-team coordination points are reflected. "
        "Your final plan should allow tracking of progress and support effective reporting to stakeholders. "
        "Use Agile metrics like velocity and team capacity as reference where applicable."
    ),
    "Customer_Proxy": ()       
}
```

```python
# Initialize the shared conversation history
conversation_history = ""

# Initial message from Customer Proxy
initial_message = (
    "I want to create a web-based mobile app for our food delivery service where my "
    "customers can place their orders online, pay online, track their orders, rate and review our "
    "service, and receive promotional offers from our business."
)

# Add the initial message to the conversation history
conversation_history += f"Customer_Proxy: {initial_message}\n"
last_message = initial_message
message_count = 1

# Agents in the conversation loop
agents = [
    business_analyst_agent,        # Refines backlog into detailed user stories
    designer_agent,                # Creates wireframes and prototypes
    architect_agent,               # Defines architecture and components
    developer_agent,               # Implements features
    security_engineer_agent,       # Reviews security of implementation
    devops_engineer_agent,         # Sets up deployment pipelines and environments
    qa_engineer_agent,             # Writes and executes test cases
    technical_writer_agent,        # Updates user guides and docs
    scrum_master_agent,            # Facilitates sprint planning and team alignment
    project_manager_agent,         # Product Owner translates vision into backlog
    customer_proxy_agent           # Loops back for feedback
]


# Set the maximum number of messages and the message count
max_messages = 30
exit_loop = False  # Flag to control breaking out of both loops

# Start the conversation loop
while message_count < max_messages and not exit_loop:
    for agent in agents:
        user_input = input("Press 'Enter' to continue the conversation between the group, or type 'exit' to end it: ")
        if user_input.lower() == 'exit':
            exit_loop = True
            break  # Breaks out of the inner loop

        # Project Manager sends the specific prompt to the agent
        prompt = project_manager_prompts[agent.name]
        # Project Manager sends the prompt as a message
        pm_message = prompt
        conversation_history += f"{project_manager_agent.name}: {pm_message}\n"

        # Agent responds to the last message with access to the conversation history
        agent_response = agent.respond(
            input_text=last_message,
            conversation_history=conversation_history
        )

        # Print the agent's response
        print(f"{agent.name}:\n{agent_response}\n")

        # Update the conversation history
        conversation_history += f"{agent.name}: {agent_response}\n"
        # Update the last message
        last_message = agent_response
        message_count += 1

        if message_count >= max_messages or exit_loop:
            break  # Breaks out of the inner loop if conditions are met

# Exit the program after breaking out of both loops
print("Conversation ended.")
```
